```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sbn
```

```python
atletas = pd.read_csv("athlete_events.csv")
regiones = pd.read_csv("regions.csv")
```

```python
atletas.head()
```

```python
#Atletas que han participado en mas de un deporte
nombres = atletas.groupby("Name").count()["Event"]
print(nombres)
```

```python
for i in range(0,len(nombres)):
    atleta = atletas.groupby("Name").count().iloc[i]["Event"]
    if atleta > 1:
        print(atletas.groupby("Name").count().iloc[i])
```

```python
#Los deportitas que han estado en ms de un evento y han ganado ms de una medalla
for i in range(0,len(nombres)):
    atleta = atletas.groupby("Name").count().iloc[i]["Event"]
    if atleta > 1:
        medallas = atletas.groupby("Name").count().iloc[i]["Medal"]
        if medallas > 1: 
            print(atletas.groupby("Name").count().iloc[i])
```

```python
#Cantidad de medallas por atleta
for i in range(0,len(nombres)):
    atleta = atletas.groupby("Name").count().iloc[i]["Medal"]
    print(atleta)
```

```python
#Que regiones han ganado ms medallas top 10
unidas = atletas.merge(regiones,left_on='NOC',right_on='NOC')
medallas = unidas[["region","Medal"]].groupby("region").count().nlargest(10,"Medal")
```

```python
unidas
```

```python
medallas
```

```python
#Que regiones han ganado ms medallas de bronce top 10
totalBronce = unidas[unidas.Medal == "Bronze"]
totalBronce[["region","Medal"]].groupby("region").count().nlargest(10,"Medal")
```

```python
#Que pas del numero total de competidores que ha enviado ha ganado la mayor cantidad de medallas
# medallas / no. jugadores *100
eficiencia = unidas[["region","Name","Medal"]].groupby("region").count()
eficiencia["Eficiencia"] = (eficiencia.Medal/eficiencia.Name)*100
eficiencia.nlargest(10,"Eficiencia")
```

```python
#Qu pais tiene el promedio de edad ms grande
edad_promedio = unidas[["region","Age"]].groupby("region").mean().nlargest(10,"Age")
#edad_promedio.nlargest(10,"Edad_promedio")
edad_promedio
```

```python
#Por cada tipo de medalla quien es el mximo y el minimo de peso y estatura
#Dentro de la funcin agg va un diccionario con llave-valor
unidas.groupby("region").agg({"Height":["min","max"],"Weight":["min","max"]})
```

```python
#Por tipo de medalla se pueda ver la edad
edad_Medallas = unidas.groupby("Medal").agg({"Age":["min","max","mean","var"]})
edad_Medallas
```

```python
#Por tipo de medalla las caracteristicas de la edad con un boxplot
plt.figure(figsize=(25,18))
sbn.boxplot(x="Medal",y="Age",data=unidas)
```

```python
#En cada deporte cuantos hombres y cuantas mujeres hay 
plt.figure(figsize=(150,100))
Fem = unidas[unidas.Sex == "F"]
Fem[["Sport","Sex"]].groupby("Sport").count().plot(kind="bar")
```

```python
plt.figure(figsize=(150,100))
Masc = unidas[unidas.Sex == "M"]
Masc[["Sport","Sex"]].groupby("Sport").count().plot(kind="bar")
```
