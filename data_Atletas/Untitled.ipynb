```python
import pandas as pd
```

```python
atletas = pd.read_csv("athlete_events.csv")
regiones = pd.read_csv("regions.csv")
```

```python
atletas.head()
```

```python
regiones
```

```python
#Cuantas medallas ha ganado México
#La función merge cruza datos de una tabla con otra
atletas.merge(regiones,left_on='NOC',right_on='NOC')
```

```python
#Medallas de oro
medallas = atletas[atletas.Medal=="Gold"]
```

```python
#Regiones que han ganado medallas de oro 
medallas.merge(regiones,left_on='NOC',right_on='NOC').groupby("region").count()
```

```python
#Paises que no han ganado ninguna medalla 
medallas = atletas[atletas.Medal.notnull()==False]
medallas.merge(regiones,left_on='NOC',right_on='NOC').groupby("region").count()
```
