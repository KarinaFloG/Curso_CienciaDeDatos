```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sbn
import shapefile as shp
```

```python
pendientes = pd.read_csv("pendientes_2020_2.csv",encoding="ISO-8859-1")
pendientes["fecha"] = pd.to_datetime(pendientes["fecha"], format="%Y-%m-%d")
maximos = pd.read_csv("max_2020_2.csv",encoding="ISO-8859-1")
```

```python
#Se elimina el valor que no sirve
maximos = maximos.drop("Unnamed: 0",axis='columns')
maximos["tz_time"] = pd.to_datetime(maximos["tz_time"], format="%Y-%m-%d")
maximos
```

```python
pendientes
```

```python
pend_max = pendientes.merge(maximos,left_on='fecha',right_on='tz_time')
#pend_max = pd.concat([maximos, pendientes], axis=1,t=False)
pend_max
```

```python
plt.plot(pend_max["pendiente_max"],color="green")
```

```python
print("Máximo pendiente_max: ",pend_max["pendiente_max"].max())
print("Minimo pendiente_max: ",pend_max["pendiente_max"].min())
print("Media pendiente_max: ",pend_max["pendiente_max"].mean())
```

```python
plt.plot(pend_max["page_uniques"],color="orange")
```

```python
print("Máximo page_uniques: ",pend_max["page_uniques"].max())
print("Minimo page_uniques: ",pend_max["page_uniques"].min())
print("Media page_uniques: ",pend_max["page_uniques"].mean())
```

```python
from scipy.signal import savgol_filter
plt.figure(figsize=(18,8))
plt.plot(pend_max["fecha"],pend_max["page_uniques"])
plt.plot(pend_max["fecha"],savgol_filter(pend_max["page_uniques"],19,1),color='red',label="Suavizado")
```

```python
from sklearn.linear_model import LinearRegression
#X debe ser en un rango en el que este page uniques y debe ser siempre un arreglo numerico
x = np.arange(len(pend_max["page_uniques"])).reshape(-1,1)
y = np.array(savgol_filter(pend_max["page_uniques"],19,1))
reg_lin = LinearRegression().fit(x,y)
#Ordenada al origen
b = reg_lin.intercept_
#Pendiente
m = reg_lin.coef_
```

```python
m
```

```python
plt.plot(pend_max["fecha"],savgol_filter(pend_max["page_uniques"],19,1),color="red",label="Suavizado")
plt.plot(pend_max["fecha"],np.arange(len(pend_max["fecha"]))*m+b)
```

```python
#Análisis hasta marzo
plt.figure(figsize=(18,8))
x = np.arange(len(pend_max["fecha"][:60])).reshape(-1,1)
y = np.array(savgol_filter(pend_max["page_uniques"][:60],19,1))
reg_lin = LinearRegression().fit(x,y)
#Ordenada al origen
b = reg_lin.intercept_
#Pendiente
m = reg_lin.coef_
plt.plot(pend_max["fecha"],savgol_filter(pend_max["page_uniques"],19,1),color="red",label="Suavizado")
plt.plot(pend_max["fecha"][:61],np.arange(len(pend_max["fecha"][:61]))*m+b)
```
